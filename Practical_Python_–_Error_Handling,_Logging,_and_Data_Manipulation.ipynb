| ----------- | ------------------- | ----------------- |

2.  What are the challenges associated with memory management in Python?
    - The challenges associated with memory management in Python are given below:
      - Python primarily tracks objects using reference counting.
      - Python has a cyclic garbage collector (GC) to clean up circular references.
      - Even after objects are deleted, memory may not be returned to the OS.
      - The GIL affects memory and performance in multithreaded programs.



```python
# 3. Write a Python program that logs an error message to a log file when a division by zero exception occurs.

# Answer

import logging

# Configure logging
logging.basicConfig(
    filename='error.log',          # Log file name
    level=logging.ERROR,           # Log only errors and above
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def divide_numbers(a, b):
    try:
        result = a / b
        print("Result:", result)
    except ZeroDivisionError as e:
        logging.error("Division by zero occurred", exc_info=True)
        print("Error: Cannot divide by zero.")

# Example usage
num1 = 10
num2 = 0

divide_numbers(num1, num2)



# 4. Write a Python program that reads from one file and writes its content to another file.

# Answer

# Copy contents from one file to another

source_file = "input.txt"
destination_file = "output.txt"

try:
    with open(source_file, "r") as src:
        content = src.read()   # Read entire file

    with open(destination_file, "w") as dest:
        dest.write(content)   # Write to new file

    print("File copied successfully!")

except FileNotFoundError:
    print("Error: Source file not found.")
except IOError:
    print("Error: Problem reading or writing file.")





# 5. Write a program that handles both IndexError and KeyError using a try-except block.

# Answer

# Example data
my_list = [10, 20, 30]
my_dict = {"a": 1, "b": 2}

try:
    # Accessing list element (may raise IndexError)
    print("List value:", my_list[5])

    # Accessing dictionary key (may raise KeyError)
    print("Dictionary value:", my_dict["c"])

except IndexError:
    print("Error: List index is out of range.")

except KeyError:
    print("Error: Key not found in dictionary.")
```
